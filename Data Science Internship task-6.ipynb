**Problem Statement 1:Smart Parking Lot Management System**


```python
def parking_management(capacity,vehicle_logs):
    current_parked =0
    peak_usage=0
    for log in vehicle_logs:
        if log == "IN":
            current_parked +=1
        elif log=="OUT":
            if current_parked>0:
                current_parked -=1
        #for peak usage
        peak_usage = max(peak_usage,current_parked)

        #alert if exceeded capacity
        if current_parked>capacity:
            print("Warning Parking capacity exceeded!")
    print("Currently Parked Vehicles:",current_parked)
    #parking status
    if current_parked<capacity:
        print("Parking Status: Available")
    else:
        print("Parking Status: Full")
    print("peak Parking Usage:",peak_usage)
    
# user input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
parking_management(capacity,vehicle_logs)
    
```

**Problem Statement 2:Online Food Delivery Time Estimator**

```python
def delivery_time_estimator(distance,traffic_level,weather):
#assume 20km/hr -->20km/60min-->1km = 3min so distance*3
    time = distance*3
#assume add 5min if traffic low,medium add 10min,high add 15min
    if traffic_level=="Low":
        time +=5
    elif traffic_level=="Medium":
        time += 10
    elif traffic_level=="High":
        time+=15
# add 11 if weather rainy,add 20min if its strom
    if weather=="Clear":
        time+=0
    elif weather=="Rainy":
        time +=11
    elif weather=="Strom":
        time += 20
    print(f"Estimated Delivery Time: {time} minutes")
#user inputs
distance = int(input("Distance (km):"))
traffic_level = input("Traffic Level: ")
weather = input("Weather: ")
#call function
delivery_time_estimator(distance,traffic_level,weather)
```

**Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

```python
def analyze_seat_booking(total_seats, booked_seats):
    
    booked_count = len(booked_seats)
    
    # Calculate occupancy for getting percentage
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", int(occupancy), "%")
    
    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
    
    # Suggest additional show
    if occupancy >= 90:
        print("Suggestion: Open Additional Show")
    else:
        print("Suggestion: No Additional Show Needed")


# user Input
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

analyze_seat_booking(total_seats, booked_seats)
```

**Problem Statement 4:Cloud Server Load Classification System**

```python
def cloud_server(cpu_readings):
     # Initialize variable to store total CPU usage
    total =0
     # Loop through each CPU reading in the list
    for reading in cpu_readings:
        total += reading
    # Calculate average CPU load
    # Using integer division (//) to remove decimal part
    avg = total//len(cpu_readings)
    # Check server status based on average CPU load
    if avg>=80:
        status= "Critical"
    elif avg >=50 and avg<80:
        status = "Warning"
    else:
        status = "Normal"
    return avg,status
#user inputs
cpu_readings= [45, 60, 70, 85, 90]
avg,status = cloud_server(cpu_readings)
print(f"Average CPU Load: {avg}%") 
print("Server Status:",status)
        
```

**Problem Statement 5:Smart Classroom Resource Usage Monitor**

```python
def track_resource_usage(resource_usage):
    
    overused = []
    
    # Check each resource
    for resource, hours in resource_usage.items():
        
        # Condition for overuse (more than 8 hours)
        if hours > 8:
            overused.append(resource)
    
    # Display results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

track_resource_usage(resource_usage)
```

**Problem Statement 6: Online Event Registration Capacity Controller**

```python
def event_registration_controller(capacity,registrations):
     # Check if total registrations are less than event capacity
    if capacity>registrations:
        # All registrations can be confirmed
        confirm_reg = registrations
        # No one is waitlisted

        wait_list = 0
        # Registration is still open
        reg_status = "Open"
    else :
        # Only capacity number of users can be confirme
        confirm_reg = capacity
        # Extra users go to waitlist
        wait_list = registrations - capacity
         # Registration is closed (capacity reached or exceeded)
        reg_status = "Closed"
    return confirm_reg,wait_list,reg_status
#user inputs
capacity = int(input("Event Capacity: "))
registrations = int(input("Registrations: "))
confirm_reg,wait_list,reg_status = event_registration_controller(capacity,registrations)
print(f"Confirmed Registrations: {confirm_reg}")
print(f"Waitlisted Users: {wait_list}")
print(f"Registration Status: {reg_status}")
```
