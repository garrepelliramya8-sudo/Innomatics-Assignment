Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
# Find highest score
highest = max(employees.values())
print("Top Performers Eligible for Bonus:",end=" ")
# Print employees who have highest score
for name in employees:
    if employees[name] == highest:
        print(name,end = ",")
print("(Score:",highest,")")    
    
```

Problem Statement 2: Search Query Keyword Analysis


```python
Sentence = input()
str = Sentence.lower().split()
dict = {}
# Count frequency
for w in str:
    if w in dict:
        dict[w] +=1
    else:
        dict[w] = 1
# Create new dictionary for repeated words
repeated = {}

for w in dict:
    if dict[w] > 1:
        repeated[w] = dict[w]

print(repeated)

```

Problem Statement 3: Sensor Data Validation


```python
sensor_readings = list(map(int,input().split(",")))
ans=[]
# if readings divisible by 2 add it and its index to list
for i in range(len(sensor_readings)):
    if sensor_readings[i]%2==0:
        ans.append((i,sensor_readings[i]))
print("Valid Sensor Readings (Hour, Value):")
print(ans)
    
    
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]   # get domain part
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total = len(emails)

# Calculate percentage
for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain + ":", int(percentage), "%")
```

Problem Statement 5: Sales Spike Detection


```python
sales =  [1200, 1500, 900, 2200, 1400, 3000]
#Calculate average
avg = sum(sales)/len(sales)
# Check for spike (30% above average) and Print days with spike
for i in range(len(sales)):
    if sales[i] > ((1.25)*avg):
        print(f"Day {i+1}: {sales[i]}") 
```
