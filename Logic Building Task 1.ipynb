```python
# 1.check whether login is successful if username="admin" and password="1234"
username = input()
password=input()

#if given username and password matches with inputs then "login successful" implements else "Invalid credentials implements"
if username=="admin" and password=="1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2. Pass/Fail Analyzer

marks = [45,78,90,3,60]
''' if marks >=50 student pases hence increment passstudents variable 
else increment failstudents variable and to get every student marks from list use for loop'''
passed = 0;
fail = 0;
for i in marks:
    if(i>=50):
        passed +=1;
    else:
        fail += 1;
print("Number of students passed are : ",passed)
print("Number of students failed are : ",fail)
        
```

```python
# 3.Simple Data Cleaner
''' we have to print the list objects without any front and back spaces and
objects should be in lowercase'''

names=[" Alice ","bob","CHARLIE "]

for i in range(len(names)):
    names[i]=names[i].strip().lower()
print(names)
    
    
```

```python
#4. Message Length Analyzer
''' length of each string in list and if string length > 10 keep flagged else print string length

messages = ["Hi","Welcome to the platform","OK"]

for i in messages:
    if len(i)>10:
        print("flagged: ",i)
    else:
        print(len(i))
```

```python
#5.Error Message Detector
''' we have to count no of "ERROR" msgs'''
logs = ["INFO","ERROR","WARNING","ERROR"]
errors =0

for i in logs:
    if i=="ERROR" :
        errors += 1
print("The no of ERROR strings in list are : ",errors)
        
```
