```python
#1. unique words in a sentence

sentence = input("Enter sentence: ")
# converting sentence to words
words = list(sentence.split())
list=[]
# checking every word in wordslist if there and
#we are checking if that word not in unique list adding that word to unique list
for w in words:
    if w not in list:
        list.append(w)

#printing no.of unique words in sentence
print("Unique words count: ",len(list))
#printing unique words in sentence
print("Unique words: ",list)
    

    
    


```

```python
# 2 .Highest Salary from Employee Data

employees = {"Ravi":75000,"Anita":68000,"Kiran":72000}
# we have to find max(high) salary so took max=0
max=0
# we have to find which employee get high salary to store it we took emp_name
emp_name=""
# as name and salary stored in dictionary took for loop to iterate over it
for name in employees:
# if employee salary greater that max store salary as marks and employee name
    if employees[name] > max:
        max=employees[name]
        emp_name=name
#printing highest salary man name with his salary
print("Highest Salary: ",emp_name,"-",max)
```

```python
#3.Find Maximum and Minimum Values
numbers = list(map(int,input("List: ").split(",")))
#assign 0 to max and first element of list as min
min=numbers[0]
max=0

# to consider every element in list using for loop
for num in numbers:
# if num greater than max then store it as max else if num lessthan min store num as min
    if num>max:
        max=num
    elif num<min:
        min = num
print("Max:",max)
print("Min:",min)
```

```python
# 4.Count Products Above a Price Threshold

prices = list(map(int,input("prices = ").split(",")))
# for storing  no.of prices > 1000
count=0

# if price greater than 1000 increment count value by1
for price in prices:
    if price>1000:
        count +=1

print("Products above 1000:",count)
```

```python
# Problem 5: Calculate Attendance Percentage
# storing attendence in list
attendence = list(map(str,input("attendence = ").split(",")))
present=0
# present variable to count no.of presentiees

# to get presentiees number if stud == presnt increment present by 1
for stud in attendence:
    if stud=="P":
        present+=1

# printing presentees percentage
print("Attendance Percentage:",((present/len(attendence))*100))
```

```python
#Problem 6: Remove Duplicate Phone Numbers
phonenum = list(map(int,input("phone_numbers = ").split(",")))
# took an empty set
s = set()
# storing every value of given list to set
for num in phonenum:
    s.add(num)
#printing set
print("Unique phone numbers:",s)
```

```python
# Problem 7: Count Character Frequency
str = input()
d={}
# for every charcter in string
# if char in dictionary increment its value by 1 else make its value as 1
for ch in str:
    if ch in d:
        d[ch]+=1
    else :
        d[ch]=1
print(d)
```
