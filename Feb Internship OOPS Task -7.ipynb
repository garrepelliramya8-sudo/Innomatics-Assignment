**Problem 1: Smart Light Controller**

```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.is_on = False   # Initially OFF

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def display_status(self):
        if self.is_on:
            print(f"{self.light_name} is ON")
        else:
            print(f"{self.light_name} is OFF")


# Taking user input
light_name = input("Light Name: ")
action = input("Action: ")

# Creating object
light = SmartLight(light_name)

# Performing action
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

# Display result
light.display_status()
```

**Problem 2: Employee ID Card System**

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Taking user input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# Creating Employee object
employee = Employee(name, emp_id, department)

# Displaying ID card
employee.display_id_card()
```

**Problem 3: Mobile Contact Record**

**Problem 4: Product Price Tag Generator**

**Problem 5: Movie Rating Display System**

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")


# Taking user input
movie_name = input("Movie Name: ")
rating = float(input("Rating: "))

# Creating Movie object
movie = Movie(movie_name, rating)

# Displaying movie details
movie.display_details()
```

**Problem 6: Delivery Address Manager**

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# Taking user input
customer_name = input("Customer Name: ")
address = input("Address: ")

# Creating Delivery object
delivery = Delivery(customer_name, address)

# Displaying delivery details
delivery.display_delivery_details()
```
