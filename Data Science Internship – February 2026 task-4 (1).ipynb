**Problem 1: Mobile Recharge Validation System**

```python
# Function to validate the recharge amount
def validate_recharge(amount):
    # List of valid recharge plans
    valid_plans = [199, 299, 399, 599]
    
    # Check if amount is less than minimum allowed recharge
    if amount < 50:
        return "Minimum recharge amount is ₹50."
    
    # Check if entered amount is one of the valid plans
    if amount not in valid_plans:
        return "Invalid plan selected. Available plans: 199, 299, 399, 599."
    
    # If all conditions are satisfied
    return "Valid"


# Infinite loop to allow user retry until valid input is entered
while True:
    try:
        # Take recharge amount as input from user
        amount = int(input("Enter recharge amount: ₹"))
        
        # Call validation function
        result = validate_recharge(amount)
        
        # If validation is successful
        if result == "Valid":
            print("Recharge successful.")
            break   # Exit loop
        
        # If validation fails
        else:
            print(result)
            print("Please try again.\n")
    
    # Handle non-numeric input error
    except ValueError:
        print("Please enter a valid numeric amount.\n")
```

**Problem 2: Inventory Reorder Alert System**

```python
# Function to check inventory stock levels
def check_inventory(products):
    
    # Loop through each product and its stock quantity
    for stock in products:
        
        # If stock is less than 15, print reorder alert
        if products[stock] < 15:
            print(f"{stock}: Reorder Alert (Stock = {products[stock]})")
        
        # Otherwise, stock is sufficient
        else:
            print(f"{stock}: Stock OK (Stock = {products[stock]})")


# Dictionary containing products and their stock quantities
inventory = {
    "Rice": 20,
    "Wheat": 10,
    "Sugar": 25,
    "Oil": 8,
    "Salt": 15
}

# Call the function
check_inventory(inventory)
```

```python
# Function to determine whether a student passes or fails
def student_grades(lst):
    
    total = 0  # Variable to store the total marks
    
    # Loop through each mark in the list
    for marks in lst:
        total += marks   # Add each mark to total
    
    # Calculate average and check if it is >= 50
    if (total / len(lst)) >= 50:
        return "Pass"   # Student passes
    
    # If average is less than 50
    return "Fail"


# Take space-separated marks input from user
lst = list(map(int, input("Enter marks: ").split()))

# Print the result (Pass or Fail)
print(student_grades(lst))
        
    
```

**Problem 4: Cab Fare Estimator with Retry Option**

```python
# Function to calculate total fare
def fare_calculation(dist, time):
    
    # Base fare charged for every ride
    base_fare = 50
    
    # Calculate normal fare (₹12 per km)
    total = base_fare + dist * 12
    
    # Apply 25% extra charge for night time (8 PM to 3 AM)
    # Time is in 24-hour format
    if time >= 20 or time < 3:
        total = total + (0.25 * total)   # Add 25% surcharge
    
    # Display final amount
    print("Total amount is:", total)


# Take distance input from user
dist = int(input("Enter distance (in km): "))

# Take pickup time input (24-hour format)
time = int(input("Enter pickup time (24-hour format): "))

# Call the function
fare_calculation(dist, time)
```

**Problem 5: Employee Attendance Eligibility**

```python
# Function to calculate attendance eligibility
def check_attendance(attendance_list):
    
    present_count = 0   # Variable to count number of present days
    
    # Loop through attendance list
    for status in attendance_list:
        if status == "P":
            present_count += 1   # Increment if student is present
    
    # Calculate attendance percentage
    percentage = (present_count / len(attendance_list)) * 100
    
    # Check eligibility condition
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Taking input from user (space-separated P and A)
attendance_list = input("Enter attendance (P/A separated by space): ").split()

# Printing result
print(check_attendance(attendance_list))
```

**Problem 6: Password Strength Checker**

```python
# Function to check password strength
def check_password_strength(password):
    
    # Check minimum length
    if len(password) < 8:
        return "Weak Password: Minimum length should be 8 characters."
    
    has_digit = False
    has_special = False
    
    # Loop through each character in password
    for ch in password:
        
        # Check if character is a digit
        if ch.isdigit():
            has_digit = True
        
        # Check if character is one of the allowed special characters
        if ch in "@#$":
            has_special = True
    
    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password: Must contain at least one digit and one special character (@ # $)."


# Take password input from user
password = input("Enter password: ")

# Print result
print(check_password_strength(password))
```
