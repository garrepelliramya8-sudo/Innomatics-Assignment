**Problem 1: Social Media – Post Engagement Analyzer**

```python
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    # Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# inputs
likes = [200, 300, 150, 400]   # Total = 1050
analyze_engagement(likes)
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_analysis(rainfall,avg):
    required_level = 50
    if avg >= required_level:
        return "Adequate Rainfall"
    else:
        return "Inadequate Rainfall"
        
    #for calculating total sum
rainfall =list(map(int, input("Enter rainfall values:").split()))
sum=0
for num in rainfall:
    sum+=num
#for calculating  avg
avg = sum//len(rainfall)
#printing avg of rainfall
print(f"Average Rainfall: {avg}")
#calling and printing the rainfall_analysis
print("Rainfall Status:",rainfall_analysis(rainfall,avg))
    
        
```

**Problem 4: Social Media – Duplicate Account Detection**

```python
def detect_duplicate_accounts(usernames):
    unique_users = set()
    duplicate_found = False
    
    # Loop through usernames
    for user in usernames:
        if user in unique_users:
            duplicate_found = True
            break
        unique_users.add(user)
    
    # Output
    if duplicate_found:
        print("Duplicate Accounts Found: Yes")
        print("Flag accounts")
    else:
        print("Duplicate Accounts Found: No")
        print("Accounts are unique")


# inputs
users = ["ramya", "john", "alex", "ramya"]
detect_duplicate_accounts(users)
```

**Problem 6: Agriculture – Premium Crop Price Filter**

```python
def premium_crop_price_filter(crops_prices):
    # to store premium crops price
    premium_crops = []
    #to stor regular crops price
    regular_crops=[]
    # checking conditions
    for price in crops_prices:
        if price>2000:
            premium_crops.append(price)
        else:
            regular_crops.append(price)
    return premium_crops
#user inputs
crops_prices = [2500,3200,2000,1500,1000]
print("Premium crops:",premium_crop_price_filter(crops_prices))



            
        
            
```

**Problem 10: Smart Electricity Bill Analyzer**

```python
def calculate_bill(units):
    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example usage
units_consumed = 250
total_bill, usage_status = calculate_bill(units_consumed)

print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
